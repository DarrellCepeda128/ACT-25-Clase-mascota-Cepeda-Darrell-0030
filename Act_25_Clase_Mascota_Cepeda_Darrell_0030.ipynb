```python
class Mascota:
    def __init__(self, nombre, especie, edad):
        self.nombre = nombre
        self.especie = especie
        self.edad = edad

    # Función 1: Simula la lectura o carga de datos en el sistema
    def leerdatos(self):
        print(f"\n>> [SISTEMA]: Cargando perfil de '{self.nombre}'...")
        print(f">> [LOG]: Especie detectada: {self.especie} | Edad: {self.edad} años.")

    # Función 2: Muestra la información de forma visual
    def mostrardatos(self):
        print("\n" + "═" * 30)
        print(f"🐾 FICHA DE MASCOTA 🐾")
        print(f"  • Nombre:  {self.nombre}")
        print(f"  • Especie: {self.especie}")
        print(f"  • Edad:    {self.edad} años")
        print("═" * 30)

# --- FLUJO INTERACTIVO ---

print("=== BIENVENIDO AL REGISTRO DE MASCOTAS ===")

# Registro de la Mascota 1
print("\n--- Datos de la Mascota 1 ---")
n1 = input("Introduce el nombre: ")
e1 = input("Introduce la especie: ")
ed1 = input("Introduce la edad: ")
mascota1 = Mascota(n1, e1, ed1)

# Registro de la Mascota 2
print("\n--- Datos de la Mascota 2 ---")
n2 = input("Introduce el nombre: ")
e2 = input("Introduce la especie: ")
ed2 = input("Introduce la edad: ")
mascota2 = Mascota(n2, e2, ed2)

# Ejecución de las funciones para ambos objetos
print("\n--- PROCESANDO RESULTADOS ---")

mascota1.leerdatos()
mascota1.mostrardatos()

mascota2.leerdatos()
mascota2.mostrardatos()

print("Programa elaborado por Cepeda Darrell NL = 10")
```
